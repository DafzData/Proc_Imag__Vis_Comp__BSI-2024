**Buscando uma imagem pela URL**

```python
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
from PIL import Image

# Copie a URL de uma imagem qualquer na Internet
urllib.request.urlretrieve(
    'https://s1.static.brasilescola.uol.com.br/be/2020/10/foguete.jpg','imagem.jpg'
)

# Salvando a imagem em uma variável
imagem_RGB = plt.imread('imagem.jpg')

#Mostrar a imagem
plt.figure(1)
plt.imshow(imagem_RGB)

```

```python
# Dimensão da imagem
print(imagem_RGB.shape)
```

```python
# Altura
print(imagem_RGB.shape[0])
```

```python
# Largura
print(imagem_RGB.shape[1])
```

Convertendo para a escala de cinza

```python
# Abrir a imagem com Pillow (PIL) e converter para escala de cinza
imagem_cinza = Image.open('imagem.bmp').convert('L')  # 'L' é para converter para escala de cinza

#Mostrar a imagem
plt.figure(1)
plt.imshow(imagem_cinza, cmap='gray')
```

**CRIANDO UM HISTOGRAMA**


```python
# Converter a imagem para um array numpy
matriz = np.array(imagem_cinza)

h = np.zeros(256)
for i in range(matriz.shape[0]):
  for j in range(matriz.shape[1]):
    h[matriz[i,j]] += 1

# Plotar o gráfico
eixo_x = list(range(256))
plt.figure(2)
plt.bar(eixo_x, h)
plt.show()
```
