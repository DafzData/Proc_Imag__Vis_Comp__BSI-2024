```python
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
from skimage import io, color  # Substitui o PIL

# Baixar a imagem
urllib.request.urlretrieve(
    'https://s1.static.brasilescola.uol.com.br/be/2020/10/foguete.jpg',
    'imagem.jpg'
)

# Carregar a imagem usando skimage
imagem_RGB = io.imread('imagem.jpg')

# Mostrar a imagem
plt.figure(1)
io.imshow(imagem_RGB)  # Pode usar tanto io.imshow quanto plt.imshow
#plt.axis('off')  # Opcional: remove os eixos
#plt.show()
```

```python
# Dimenso da imagem
print(imagem_RGB.shape)
```

```python
# Altura
print(imagem_RGB.shape[0])
```

```python
# Largura
print(imagem_RGB.shape[1])
```

```python
# Carregar a imagem diretamente em escala de cinza com skimage
# imagem_cinza = io.imread('imagem.bmp', as_gray=True)
imagem_cinza = (io.imread('imagem.jpg', as_gray=True) * 255).astype('uint8')

# Mostrar a imagem
plt.figure(1)
plt.imshow(imagem_cinza, cmap='gray')
plt.axis('off')  # Remove os eixos
plt.show()
```

```python
# Converter a imagem para um array numpy
matriz = np.array(imagem_cinza)

h = np.zeros(256)
for i in range(matriz.shape[0]):
  for j in range(matriz.shape[1]):
    h[matriz[i,j]] += 1

# Plotar o grfico
eixo_x = list(range(256))
plt.figure(2)
plt.bar(eixo_x, h)
plt.show()
```
