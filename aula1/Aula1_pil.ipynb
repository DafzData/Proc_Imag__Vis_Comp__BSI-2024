**AULA 01 - Exibindo, convertendo e redimensionando uma imagem**

**EXIBINDO**

obs.: carregue uma imagem na pasta do colab ao lado (pode ser baixada do reposit√≥rio no github)

```python
# Biliotecas
import numpy as ap
from PIL import Image
import matplotlib.pyplot as plt

# Abrir a imagem
imagem = Image.open('buildings.jpg')

# Mostrar a imagem
plt.imshow(imagem)
#plt.axis('off') # para na√£o mostrar os pixels
#plt.show()
```

**CONVERTENDO para escala de cinza**

```python
# Convertter para a escala de cinza
imagem_cinza = imagem.convert('L')

# Mostrar a imagem em escala de cinza
plt.imshow(imagem_cinza, cmap='gray')
#plt.show()
```

```python
# Obter largura (width) e altura (height) da imagem
width, height = imagem_cinza.size

print(width) # largura
print(height) # altura
```

**REDIMENSIONANDO A IMAGEM**

```python
# Definir nova largura e altura da imagem
new_width = int(width/20) # reduzir ou aumentar quantas vezes quiser
new_height = int(height/20)

# Redimensionar a imagem
imagem_redimensionada = imagem_cinza.resize((new_width, new_height))

# Mostrar a imagem
plt.imshow(imagem_redimensionada)
plt.imshow(imagem_redimensionada, cmap='gray')
#plt.show()
```

```python
# Obter largura (width) e altura (height) da NOVA imagem
width, height = imagem_redimensionada.size

print(width) # largura
print(height) # altura
```
