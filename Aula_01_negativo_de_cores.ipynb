# Transformação de Imagem: Criando um Negativo de Cores



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carrega a imagem (substitua pelo caminho da imagem ou use uma exemplo)
imagem = cv2.imread("/content/cao_z.jpg")  # ou use np.zeros((100, 100, 3)) para criar uma imagem preta

# Converte de BGR (padrão OpenCV) para RGB
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# Inverte os canais de cor (negativo)
imagem_negativa = 255 - imagem_rgb

# Exibe as imagens
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagem_negativa)
plt.title("Negativa")
plt.axis('off')

plt.show()
```
